# Vegetation in Saint Martin Island

During our undergrad, department of Geography and Environment, University of Dhaka, arranged field trip and visited Teknaf and Saint Martin Island.

## Introduction

Vegetation survey is an important research agenda in vegetation science. It defines vegetation types and helps understand differences among them, which is essential for both basic ecological research and applications in biodiversity conservation and environmental monitoring.

First we selected 50 sq. m in an area. Then we selected 25 points on both side in the area. We observed the whole area & evaluated the whole Saint Martin’s island. Then we took notes about the tree names and sample of the trees. Then we went to the Botany laboratory of Department of Botany, University of Dhaka & found out the scientific names of the trees. Then we found out the Importance Value Index (IVI) of the trees. Most of the trees were Keya trees and Shagorlota trees.

Important Value Index (IVI) = RD + RDo + FR where:

- RD= Relative density
- RDo= Relative dominance
- FR= Frequency relative

```python
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
data=pd.read_csv("vegetation.csv")
```

```python
display(data)
```

```python
labels=[]
percent_data=data[['IVI']]/data[['IVI']].sum()
percent_data=percent_data.values[:,0].tolist()
species=data['Individual  species'].values.tolist()

for i in range(len(species)):
    lbl=species[i]+" "+str(round(percent_data[i]*100,2))+"%"
    labels.append(lbl)

plt.pie(percent_data, autopct='%1.1f%%',radius=2.5)
plt.legend( labels,title="IVI",loc="right",ncol=1,bbox_to_anchor=(3,0.5))
#plt.legend(bbox_to_anchor=(0, 1), loc='upper left', ncol=1)
```

### Distribution of Important Value Index
The picture aboved present the distribution of Important Value Index (IVI) of different species of tree. We can clearly see that chhotonfuti ful has the higest IVI value

```python
labels=[]
percent_data=data[['Relative  frequency']]/data[['Relative  frequency']].sum()
percent_data=percent_data.values[:,0].tolist()
species=data['Individual  species'].values.tolist()

for i in range(len(species)):
    lbl=species[i]+" "+str(round(percent_data[i]*100,2))+"%"
    labels.append(lbl)


plt.pie(percent_data, autopct='%1.1f%%',radius=2.5)
plt.legend( labels,title="Frequency",loc="right",ncol=1,bbox_to_anchor=(3,0.5))
#plt.legend(bbox_to_anchor=(0, 1), loc='upper left', ncol=1)
```

### Distribution of Relative Frequency
